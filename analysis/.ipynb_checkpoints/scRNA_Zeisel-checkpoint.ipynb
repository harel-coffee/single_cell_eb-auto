** pca plot **

```python
n_components=1000
pca=PCA(n_components=n_components)
pca.fit(X)
X_tran=pca.transform(X)
```

```python
## explained variance
temp=np.zeros(500)
temp[0]=pca.explained_variance_ratio_[0]
for i in range(1,500):
    temp[i]=temp[i-1]+pca.explained_variance_ratio_[i]
plt.figure()
plt.plot(temp)
plt.xlabel('# components')
plt.ylabel('percetage of explained variance ')
plt.show()
```

```python
## sparsity of the components: seems the components are not sparse 
Q=pca.components_
plt.figure()
for i in range(5):
    plt.plot(np.sort(np.absolute(Q[i,:]))[-300:],label=str(i))
plt.ylim([0,0.15])
plt.legend()
plt.xlabel('rank of the entry')
plt.ylabel('magnitude')
plt.show()
```

```python
## try to generate a heat map 
idx_select=[]
for i in range(100):
    temp=np.argsort(np.absolute(Q[i,:]))[-2:]
    for j in temp:
        if j not in idx_select:
            idx_select.append(j)
plt.figure()
plt.imshow(np.absolute(Q[:,idx_select]), cmap='hot', interpolation='nearest')
plt.show()
```

```python
## go for the refined structure 
X_select=X_label['group #']==3
X_label_=X_label['group #'][X_select]
X_=X[X_select,:]
feature_select=((np.sum(X_>0,axis=0)/(X_.shape[0]+0.0))>0.8)*np.std(X_,axis=0)>0.3
print('#features:%s'%str(np.sum(feature_select)))
X_=X_[:,feature_select]
pca=PCA(n_components=100)
pca.fit(X_)
X_tran_=pca.transform(X_)
plot_scatter(X_tran_,X_label_,0,1)
```
