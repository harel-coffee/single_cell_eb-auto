```python
def dd_1d_beta(M,verbose=False):
    theta_0 = np.array([1,10,5,20,0.5])
    x = np.linspace(1e-4,1-1e-4,101)
    cons = ({'type': 'ineq', 'fun': lambda x:  x},{'type': 'ineq', 'fun': lambda x: 1-x[4]})
    res = sp.optimize.minimize(f_opt,theta_0,args=(x,M),jac=False,constraints=cons,options={'disp': True})
    a1,b1,a2,b2,p = res.x
    
    p1 = beta.pdf(x,a1,b1)
    p2 = beta.pdf(x,a2,b2)      
    p1 = p1/np.sum(p1)
    p2 = p2/np.sum(p2)
    p_hat = p*p1+(1-p)*p2
    
    return p_hat,x      
```

```python
from scipy.stats import beta
def f_cal(theta,x,M):
    a1,b1,a2,b2,p = theta
    p1 = beta.pdf(x,a1,b1)
    p2 = beta.pdf(x,a2,b2)      
    p1 = p1/np.sum(p1)
    p2 = p2/np.sum(p2)
    K = M.shape[0]
    M_hat = np.zeros([K],dtype=float)
    for k in range(K):
        M_hat[k] = p*p1.dot(x**(k+1)) + (1-p)*p2.dot(x**(k+1))
    return np.sum((M_hat-M)**2)
```
