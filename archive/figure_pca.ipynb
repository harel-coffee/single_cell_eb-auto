```python
def plot_PCA_score(data,v1,v2,label=None,n_cell=None,ref_score=None,return_score=False):
    if n_cell is not None:
        subsample_index = np.random.choice(np.arange(data.shape[0]),n_cell,replace=False)  
    X=data.X
    score1 = X.dot(v1)
    score2 = X.dot(v2)
    
    if ref_score is not None:
        ref_score1 = ref_score[0]
        ref_score2 = ref_score[1]
        if np.linalg.norm(ref_score1-score1)>np.linalg.norm(ref_score1+score1):
            score1 = -score1
        if np.linalg.norm(ref_score2-score2)>np.linalg.norm(ref_score2+score2):
            score2 = -score2
    
    plt.scatter(score1,score2,s=4,alpha=0.3)
    if return_score:
        return [score1,score2]
    
def preprocess(data):
    data_ = data.copy()
    A = data_.X.data
    cap = np.percentile(A,99)
    A = A.clip(max=cap)
    data_.X.data = A
    return data_
```

```python
## pca on the original data
size_factor_full = sd.dd_size_factor(data,verbose=False)
mean_full,cov_full,PC_full  = sd.ml_covariance(data,size_factor=None)
D_full,V_full = np.linalg.eigh(cov_full)

v1 = np.array(V_full[:,-2]).reshape(-1)
v2 = np.array(V_full[:,-3]).reshape(-1)
plt.figure()
plot_PCA_score(data,v1,v2)
plt.show()
```

```python
def M_to_cv(M):
    var_ = (M[1]-M[0]**2).clip(min=1e-6)
    cv_ = np.sqrt(var_)/M[0]
    return cv_

def get_rank(x):
    n_x = x.shape[0]
    rank = np.zeros([n_x],dtype=int)
    rank[np.argsort(x)] = np.arange(n_x)[::-1]
    return rank
```

```python
size_factor = sd.dd_size_factor(data,verbose=False)
M_ml, M_dd = sd.dd_1d_moment(data,size_factor=size_factor)
cv_dd = M_to_cv(M_dd)
cv_ml = M_to_cv(M_ml)
```
