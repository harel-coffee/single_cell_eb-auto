# Beta Poisson ML estimation 

```python
## theta=[alpha,beta]
from scipy.stats import beta
def ml_beta_poisson(Y,gamma=None):
    def f_opt_beta_poisson(theta,Y_pdf,x,P_model):
        return f_cal_beta_poisson(theta,Y_pdf,x,P_model)

    def f_cal_beta_poisson(theta,Y_pdf,x,P_model):
        a,b = theta
        gx = beta.pdf(x,a,b)
        gx = gx/np.sum(gx)
        fy = P_model.dot(gx)
        l = Y_pdf.dot(np.log(fy))
        return -l
    
    x = np.linspace(1e-4,1-1e-4,101)
    Y_pdf,Y_supp = counts2pdf_1d(Y)   
    P_model = Pmodel_cal(x,Y_supp,gamma)    
    
    cons = ({'type': 'ineq', 'fun': lambda x:  x})
    theta_0 = np.array([1,1])    
    res = sp.optimize.minimize(f_opt_beta_poisson,theta_0,args=(Y_pdf,x,P_model),\
                               jac=False,constraints=cons,options={'disp': False})
    theta = res.x
    return theta

## theta=[alpha,beta]
from scipy.stats import beta
def ml_beta_poisson_m(Y,gamma=None):
    def f_opt_beta_poisson(theta,x,M):
        return f_cal_beta_poisson(theta,x,M)

    def f_cal_beta_poisson(theta,x,M):
        M_hat=M_cal_beta(theta,x,M.shape[0])
        return np.sum(((M_hat-M)/np.sqrt(M))**2)
    
    def M_cal_beta(theta,x,K):
        a,b = theta
        px = beta.pdf(x,a,b)
        px = px/np.sum(px)
        M = np.zeros([K],dtype=float)
        for k in range(K):
            M[k] = px.dot(x**(k+1))
        return M
        
    K=4
    x = np.linspace(1e-4,1-1e-4,101)
    Y_pdf,Y_supp = counts2pdf_1d(Y)   
    P_model = Pmodel_cal(x,Y_supp,gamma)    
    _,_,M,N_r_hat = dd_moments_1d(Y,k=K,gamma=gamma,noise='poi')
    print('M_hat',M)
    
    cons = ({'type': 'ineq', 'fun': lambda x:  x})
    theta_0 = np.array([1,1])    
    res = sp.optimize.minimize(f_opt_beta_poisson,theta_0,args=(x,M),\
                               jac=False,constraints=cons,options={'disp': True})
    theta = res.x
    return theta

def em_beta_poisson(Y,gamma=None):
    def fy_cal(Y,x,theta,P_model):
        a,b = theta
        gx = beta.pdf(x,a,b)
        gx = gx/np.sum(gx)
        fy = P_model.dot(gx)
        return fy[Y]
    def px_cal(x,theta1,theta2,p):
        a1,b1 = theta1
        a2,b2 = theta2
        p1 = beta.pdf(x,a1,b1)
        p2 = beta.pdf(x,a2,b2)      
        p1 = p1/np.sum(p1)
        p2 = p2/np.sum(p2)
        return p*p1+(1-p)*p2

    n_component = 2 
    n_sample = Y.shape[0]
    n_itr = 100
    
    ## 
    x = np.linspace(1e-4,1-1e-4,101)
    Y_pdf,Y_supp = counts2pdf_1d(Y)   
    P_model = Pmodel_cal(x,Y_supp,gamma) 
    
    ## initialization 
    theta1 = np.array([1,10])  
    theta2 = np.array([4,5])  
    p = 0.5
    p_old = 0.5
    w = np.zeros([n_sample,n_component],dtype=float)
    
    
    for i in range(n_itr):
        p_old = p
        ## E_step
        w[:,0] = p*fy_cal(Y,x,theta1,P_model)
        w[:,1] = (1-p)*fy_cal(Y,x,theta2,P_model) 
        w      = (w.T/w.sum(axis=1)).T
        
        bins=np.linspace(-0.5,15.5,17)
        plt.figure(figsize=[12,5])
        plt.hist(Y,bins=bins,weights=w[:,0],label='component1',color='navy',alpha=0.6)
        plt.hist(Y,bins=bins,weights=w[:,1],label='component2',color='orange',alpha=0.6)
        plt.show()
        ## M_step
        bern = np.random.binomial(1,w[:,0])
        Y1 = Y[bern==1]
        Y2 = Y[bern==0]
        p = w[:,0].mean()
        theta1 = ml_beta_poisson(Y1,gamma=gamma)
        theta2 = ml_beta_poisson(Y2,gamma=gamma)
        
        if np.absolute(p-p_old)<1e-3: break
    
    print('theta1',theta1)
    print('theta2',theta2)
    print('p',p)
    p_hat = px_cal(x,theta1,theta2,p)
    return p_hat,x
```
