```python
def ranks(input):
    output = [0] * len(input)
    for i, x in enumerate(sorted(range(len(input)), key=lambda y: input[y])):
        output[x] = i
    return np.array(output)

def k_largest_index_argsort(a, k):
    idx = np.argsort(a.ravel())[:-k-1:-1]
    return np.column_stack(np.unravel_index(idx, a.shape))
```

```python
Rdd=np.zeros_like(PC_dd)
for i in range(len(gene_list)):
    Rdd[i,:] = ranks(PC_dd[i,:])
    
Rml=np.zeros_like(PC_ml)
for i in range(len(gene_list)):
    Rml[i,:] = ranks(PC_ml[i,:])
```

```python
Rank_diff = np.abs(Rdd - Rml)
```

```python
indi=k_largest_index_argsort(Rank_diff, 50)
```

```python
for ind in indi:
    gene = gene_list[ind[0]]
    gene2 = gene_list[ind[1]]
    print(gene,gene2)
    print('dd:',PC_dd[ind[0],ind[1]], 'rank', ,PC_dd[ind[0],ind[1]])
    print('ml:',PC_ml[ind[0],ind[1]])
                print()
```
