# Demo

```python
output_folder = './figures/figure_zero_estimator'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
np.random.seed(0)
p_ = np.array([0.2,0.5,0.3])
x_ = np.array([0,0.05,0.1])
kappa = 10
p0_true = np.sum(p_.dot(np.exp(-kappa*x_)))
n_rep = 1000
Nc=1000
t_list = np.arange(10)+0.5
n_t = len(t_list)
err = np.zeros([n_t,n_rep])
bias = np.zeros([n_t])
variance = np.zeros([n_t])
L=40

for i_t,t in enumerate(t_list):
    Nr = kappa/t
    w = sd.smooth_zero_estimator(L,t=t,restrict_t=False)
    temp_rec = np.zeros([n_rep])
    
    for i_rep in np.arange(n_rep):
        ## generate poisson data 
        x_samp = np.random.choice(x_,Nc,p=p_,replace=True)  
        Y = np.random.poisson(x_samp*Nr)
        Y = np.bincount(Y)
        Y = Y/Y.sum()
        p0_hat = w[0:Y.shape[0]].dot(Y)
        err[i_t,i_rep] = (p0_hat-p0_true)**2
        temp_rec[i_rep] = p0_hat
    bias[i_t] = temp_rec.mean() - p0_true
    variance[i_t] = temp_rec.var()

err = np.log10(err)
err_mean = err.mean(axis=1)
err_std = err.std(axis=1)

plt.figure(figsize=[10,5])

plt.subplot(121)
plt.plot(t_list,err_mean,marker='o',color='orange')
plt.fill_between(t_list,err_mean-err_std,err_mean+err_std,alpha=0.4,color='orange')
plt.xlabel('$t_c$')
plt.ylabel('log10 MMSE')

plt.subplot(122)
plt.plot(t_list,np.log10(bias**2),marker='o',label='$bias^2$')
plt.plot(t_list,np.log10(variance),marker='o',label='$var$')
plt.xlabel('$t_c$')
plt.ylabel('log10 scale')
plt.legend(fontsize=13)
plt.tight_layout()
plt.savefig(output_folder+'/error.png')
plt.savefig(output_folder+'/error.pdf')
plt.show()
```

```python
x_grid = np.linspace(0,1,51)
exp_k = np.exp(-x_grid*10)
L = 50

plt.figure(figsize=[12,4])
for i_t,t in enumerate([3,4,5]):
    plt.subplot('13'+str(i_t+1))
    plt.plot(x_grid,exp_k,label='$exp($-$\kappa x)$')

    w = sd.smooth_zero_estimator(L,t=t,restrict_t=False)
    Nr = kappa/t
    val = np.zeros(x_grid.shape[0])
    for i_x,x_ in enumerate(x_grid):
        poi_pmf = sp.stats.poisson.pmf(np.arange(L+1),Nr*x_)
        poi_pmf /= poi_pmf.sum()
        val[i_x] = w.dot(poi_pmf)
    plt.plot(x_grid,val,label='apr: t=%d'%t)
    plt.legend(fontsize=13)
    plt.xlabel('x')
    plt.ylabel('f(x)')
plt.tight_layout()
plt.savefig(output_folder+'/bias.png')
plt.savefig(output_folder+'/bias.pdf')
plt.show()
```

### Correctness check: on a toy example

```python
## testing on a toy example
data,X,p0_true,size_factor = load_toy_ann_data()
print(p0_true)
Y = np.array(data.X.todense(),dtype=int)
```

```python
w = sd.zero_component_estimator(20)
print(w)
```

```python
## sanity check
p0_ml_,p0_dd_ = sd.dd_zero_prob(data,verbose=True)
w = sd.zero_component_estimator(20)
w = np.concatenate([w,np.zeros([100])])
for i in range(3):
    p0_dd = 0
    for i_row in range(Y.shape[0]):
        p0_dd += w[Y[i_row,i]]
    p0_dd = p0_dd/Y.shape[0]
    p0_ml =np.mean(Y[:,i]==0)
    print('dd i=%d, val=%0.4f, true val=%0.4f'%(i,p0_dd_[i],p0_dd))        
    print('ml i=%d, val=%0.4f, true val=%0.4f'%(i,p0_ml_[i],p0_ml))
```

```python
zero_matrix_ml,zero_matrix_dd = sd.dd_pairwise_zero_prob(data)
print('# Sanity check')
w = sd.zero_component_estimator(20)
w = np.concatenate([w,np.zeros([100])])
for i in range(3):
    for j in range(i+1,3):
        p0_dd = 0
        p0_ml = 0
        for i_row in range(Y.shape[0]):
            p0_dd += w[Y[i_row,i]] * w[Y[i_row,j]]
            if Y[i_row,i]==0 and Y[i_row,j]==0:
                p0_ml+=1
        p0_dd = p0_dd/Y.shape[0]
        p0_ml = p0_ml/Y.shape[0]
        print('dd i=%d, j=%d, val=%0.4f, true val=%0.4f'%(i,j,zero_matrix_dd[i,j],p0_dd))        
        print('ml i=%d, j=%d, val=%0.4f, true val=%0.4f'%(i,j,zero_matrix_ml[i,j],p0_ml))
```

```python
## Mutual information
zero_matrix_ml,zero_matrix_dd = sd.dd_pairwise_zero_prob(data)
mi_dd = sd.zero_to_mi(zero_matrix_dd)
mi_ml = sd.zero_to_mi(zero_matrix_ml)
mi_true = np.zeros([3,3])
for i in range(3):
    for j in range(3):
        p0 = p0_true[i,i]
        p1 = p0_true[j,j]
        p01 = p0_true[i,j]
        mi = 0
        mi += -p0*np.log(p0.clip(min=1e-6)) - (1-p0)*np.log((1-p0).clip(min=1e-6))
        mi += -p1*np.log(p1.clip(min=1e-6)) - (1-p1)*np.log((1-p1).clip(min=1e-6))
        mi += p01*np.log(p01.clip(min=1e-6)) + (p0-p01)*np.log((p0-p01).clip(min=1e-6))\
        + (p1-p01)*np.log((p1-p01).clip(min=1e-6)) + (1-p0-p1+p01)*np.log((1-p0-p1+p01).clip(min=1e-6))
        mi_true[i,j] = mi

print('mi_true')
print(mi_true)
print('test zero_to_mi')
print(sd.zero_to_mi(p0_true))
print('mi_dd')
print(mi_dd)
print('mi_ml')
print(mi_ml)
```
